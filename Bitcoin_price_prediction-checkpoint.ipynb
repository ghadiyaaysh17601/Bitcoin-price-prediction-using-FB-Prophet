# Bitcoin Price Prediction using FBProphet


```python
#inporting libraries
import pandas as pd
from fbprophet import Prophet
from fbprophet.plot import plot
import csv
```

```python
#loading CSV File

df = pd.read_csv(r'C:\Users\Ayush Patel\Downloads\BTC-USD (2).csv')
```

```python
df.head()

```

```python
df.tail()
```

```python
df.plot()
```

```python
df = df[["Date", "Close"]]
df.columns = ["ds", "y"]
```

```python
df.head()
```

```python
df.tail()
```

```python
#initialize the model
model=Prophet()
```

```python
df.columns
```

```python
model.fit(df)
```

```python
model.component_modes
```

```python
#create future dates of 365
future = model.make_future_dataframe(periods=365)
print(future)
```

```python
forecast = model.predict(future)
```

```python
forecast.head()
```

```python
model.plot(forecast)
```

```python
model.plot_components(forecast)
```

```python
forecast[["ds", "yhat", "yhat_lower", "yhat_upper"]].tail(200)
```

```python
#putting prediction values in dataset
data=forecast[["ds", "yhat", "yhat_lower", "yhat_upper"]]
data.to_csv(r'C:\Users\Ayush Patel\.spyder-py3\prediction_dataset.csv')
```
